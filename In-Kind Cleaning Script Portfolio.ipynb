## Import libraries ##

```python
import pandas as pd
import numpy as np
```

## Upload CSV

```python
in_kind=pd.read_csv('In-Kind Tracking.csv') 
```

```python
in_kind.head(0)
```

## Drop Volunteer Name column to remove PII from data visualizations

```python
in_kind = in_kind.drop(columns=['Volunteer Name'])
```

## Verify Volunteer Name column is dropped

```python
in_kind.head(0)
```

## Define and replace Funding Names to create more standardized and readable names

```python
mapping = {
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1',
    '[REDACTED]': 'Funding Source 1'
}

in_kind['Funding Name'] = in_kind['Funding Name'].replace(mapping)
```

## Create new column (Reporting Category) to separate state-level funding categories

```python
in_kind["Reporting Category"] = np.where(
    in_kind["Program"].str.contains("[REDACTED]", case=False, na=False) &
    in_kind["Program"].str.contains("[REDACTED]", case=False, na=False),
    "[REDACTED] State",
    np.where(
        in_kind["Program"].str.contains("[REDACTED]", case=False, na=False) &
        in_kind["Program"].str.contains("[REDACTED]", case=False, na=False),
        "[REDACTED] State",
        "Other"
    )
)
```
